                                                      ** cleaning the age column**


---

```python
df_full['Age'] = df_full['Age'].map(lambda x: str(x).strip('_ ,"'))
```

```python
df_full['Age'] = df_full['Age'].astype('int64')
```

```python
df_full['Age'].dtypes
```

```python
#calculate the min and max mode: which is the min value that appears most often. and the max too

import scipy.stats as stats 
list_of_ages = df_full.groupby('Customer_ID')['Age'].apply(list)
x, y = list_of_ages.apply(lambda x: stats.mode(x)).apply([min, max])
mini, maxi = x[0][0], y[0][0]
```

```python
#create a temp column and assign NaN to every value less than 14 or bigger than 56
temp_age_column = df_full['Age'].transform(lambda x: np.NaN if ((x<mini)|(x>maxi)) else x)
```

```python
# df_full.groupby('Customer_ID')['Age'].describe()
#df_full.groupby('Customer_ID')['Age'].apply(print)
```

```python
# fill the all the ages of each customer ID with the local mode for all values, if its empty then fill it with NaN
mode_by_group = df_full.groupby('Customer_ID')['Age'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.NaN)
result = temp_age_column.fillna(mode_by_group)
```

```python
df_full['Age']=result
```
